```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, LinearSVC


df_x = df.drop(["Y"], axis=1)
df_y = df[['Y']].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(
    df_x, df_y, test_size=0.33, shuffle=True, random_state=0)
clf = SVC(decision_function_shape='ovo', kernel="rbf").fit(X_train, y_train.ravel())

clf.score(X_test, y_test)
```

```python
clf = SVC(decision_function_shape='ovo', kernel='linear').fit(X_train, y_train.ravel())

clf.score(X_test, y_test)
```

```python
clf = SVC(decision_function_shape='ovo', kernel='poly').fit(X_train, y_train.ravel())

clf.score(X_test, y_test)
```
